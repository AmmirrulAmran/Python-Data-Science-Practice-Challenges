```python
import numpy as np

np.random.seed(5555)
gold = np.random.randint(low=0, high=10, size=(7,7))

print(gold)
# [[2 3 0 5 2 0 3]
#  [8 8 0 7 1 5 3]
#  [0 1 6 2 1 4 5]
#  [4 0 8 9 9 8 7]
#  [4 2 7 0 7 2 1]
#  [9 8 9 2 5 0 8]
#  [1 9 8 2 6 4 3]]

locs = np.array([
    [0,4],
    [2,2],
    [2,3],
    [5,1],
    [6,3]
])


GoldAmount = np.array([])
for x in locs:


  xPosition =  x[0]
  yPosition = x[1]
  print(x)
  GoldAmount = np.append(GoldAmount,gold[xPosition][yPosition])


for i in range(1,6):
  print("Amount of Gold in Position" + str(i) + " is " + str(GoldAmount[i-1]))
```
