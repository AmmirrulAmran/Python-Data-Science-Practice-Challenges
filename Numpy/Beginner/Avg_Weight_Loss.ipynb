```python
import numpy as np


dailywts = 185 - np.arange(5*7)/5

print(dailywts)
#
```

```python
weightByWeek = dailywts.reshape(5,7);
print(weightByWeek)
```

```python
#Average Weight each week

weeklyAverage = np.array([])
for week in weightByWeek:
    print(week.mean())
    weeklyAverage = np.append(weeklyAverage,week.mean())




```

```python
#Average weight loss per Week

WeightLoss = np.array([])

for weeks in weightByWeek:
    dailyLoss = np.diff(weeks)
    avg_loss = dailyLoss.mean()
    WeightLoss = np.append(WeightLoss,avg_loss)

print(WeightLoss)

```
